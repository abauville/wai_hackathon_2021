```python
# Imports 
import numpy as np
import matplotlib as plt
import seaborn as sns
import matplotlib as mpl
import pandas as pd

# Figure style
plt.style.use('seaborn')
mpl.rc('font', size=13)
mpl.rc('axes', labelsize='large')
mpl.rc('xtick', labelsize='large')
mpl.rc('ytick', labelsize='large')

plt.rcParams['figure.figsize'] = [20, 10] # For larger plots
```

```python
df = pd.read_csv("./Data/SafeCity WEB+MOB_Combined - FILE2 - WEB+MOB_Combined.csv")
```

```python
df.head()
```

```python
split = df.CATEGORY.str.split(", ", expand=True)
```

```python
split_flat = pd.DataFrame(split.values.flatten())
```

```python
list_cat = split_flat[0].unique().tolist()
# list_cat.pop(1) # remove ''
# list_cat.pop(1) # remove None
list_cat
```

```python
df_cat = pd.DataFrame(np.zeros((df.shape[0],len(list_cat))),columns=list_cat,dtype=bool)
split.head()
for i in range(split.shape[0]):
    for j in range(split.shape[1]):
        cat = split.iloc[i,j]
#         if cat == None:
#             break
#         else:
        df_cat[cat][i] = True
df_cat.drop(columns=['',None])
```

```python
df = df.join(df_cat,rsuffix='CAT_')
# pd.merge(df,df_cat,suffixes=['', 'CAT_'])
df.drop(columns=['CATEGORY'])
```

```python
split_loc = df.LOCATION.str.split(r' ')    
df['COUNTRY'] = ''
temp_country = pd.Series(np.zeros(df.shape[0]))
for i in range(df.shape[0]):
    temp_country[i] = split_loc[i][-1] 
# df['COUNTRY'] = temp_country
temp_country
```
