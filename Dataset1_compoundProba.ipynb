```python
# Imports 
import numpy as np
import matplotlib as plt
import seaborn as sns
import matplotlib as mpl
import pandas as pd

# Figure style
plt.style.use('seaborn')
mpl.rc('font', size=13)
mpl.rc('axes', labelsize='large')
mpl.rc('xtick', labelsize='large')
mpl.rc('ytick', labelsize='large')

plt.rcParams['figure.figsize'] = [20, 10] # For larger plots

# Load data
df = pd.read_csv("./Data/20200306 Data International Women's Day Viz5 Launch.csv")
```

```python
df_A = df[df['Country']=='Afghanistan']
df_A = df_A[df_A['Question']=='... for at least one specific reason']
df_A.drop(columns=['Question', 'Survey Year'])
df.Country.unique()
```

```python
df_A_age = df_A[df_A['Demographics Question']=='Age']
df_A_edu = df_A[df_A['Demographics Question']=='Education']
```

```python
df_A_age_f = df_A_age[df_A_age['Gender']=='F']
df_A_edu_f = df_A_edu[df_A_edu['Gender']=='F']
```

```python
age_vs_edu = np.array([df_A_age_f['Value'].values/100.0]).T @ np.array([df_A_edu_f['Value'].values/100.0])
df_age_vs_edu = pd.DataFrame(age_vs_edu, index=df_A_age_f['Demographics Response'], columns=df_A_edu_f['Demographics Response'])
```

```python
sns.heatmap(data=df_age_vs_edu)
```
